```python
def TurnToList(tags):
    alltags2=tags.replace("'","")
    alltags2=alltags2.replace("[","")
    alltags2=alltags2.replace("]","")
    alltags2=alltags2.split(",")
    alltags3=list(map(lambda x:x.strip(),alltags2))

    return alltags3
```

```python
TurnToList(xv_10["Tags"][200])  ## list
```

```python
xv_10["Tags"] = xv_10["Tags"].apply(lambda x: TurnToList(x))
```

```python
def fn(x, y):
	if x is None:
		x=[]
	elif y is None:
		y=[]
	else:
		return x+y
```

```python
existing_tags=set(reduce(fn,xv_10["Tags"] ))
existing_tags
```

```python
for et in existing_tags:
    xv_10[et] = np.nan
```

```python
# Correlation with target variable (Largest and smallest 15)

import os
print(os.getcwd())

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
## Means that I don't need to call plt.show() to show the plots 
```

```python
path = "./dataset/application_train.csv"

app_train = pd.read_csv(path)
app_train.head(2)
```

```python
'''
自 20 到 70 歲，切 11 個點，進行分群比較 (KDE plot)
以年齡區間為 x, target 為 y 繪製 barplot
'''

app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
age_data = app_train[['TARGET','DAYS_BIRTH']]
age_data['YEARS_BIRTH'] = age_data['DAYS_BIRTH']/365
```

```python
age_data
```

```python
bin_cut = np.arange(20,75,5)
print(type(bin_cut))
age_data['YEARS_BINNED'] = pd.cut(age_data['YEARS_BIRTH'], bins = bin_cut) 
```

```python
# 顯示不同組的數量
age_data=age_data.sort_values('YEARS_BINNED',ascending=True)
len(age_data)
```

```python
plt.figure(figsize = (8, 8))
px = age_data['YEARS_BINNED']
py = age_data['TARGET']
sns.barplot(px, py,estimator=sum)
```
